PREPARATION OF DATASET

```python
# Predictor contains independant variables
# Target contains dependant variables

predictor = df.iloc[:,df.columns!= "Drug Resistance"]
Target = df.iloc[:,df.columns == "Drug Resistance"]
```

 SPLIT THE DATASET INTO TRAIN AND TEST 

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(predictor, Target, random_state = 1, test_size = 0.20)
print("x_train", x_train.shape)
print("y_train", y_train.shape)
print("x_test", x_test.shape)
print("y_test", y_test.shape)
```

TRAINING THE MODEL USING LOGISTIC REGRESSION, #when the dependant variable is dichotomous (YES or NO) / binary

```python
from sklearn.linear_model import LogisticRegression

#Create an instance of Logistic Regression
logr = LogisticRegression()

#Fit the data to the model
logr.fit(x_train, y_train)

#test results using test data
y_pred = logr.predict(x_test)
```

```python
y_pred
```

CHECK PERFORMANCE OF THE CLASSIFICATION MODEL USING CONFUSION MATRIX (which helps to see the number of correct or incorrect predictions)

```python
#Evaluate model, by passing the y_pred array and comparing it to the y_test array
from sklearn import metrics
cnf_matrix = metrics.confusion_matrix(y_pred, y_test)
cnf_matrix
```

```python
# the 1st diagonal quadrants tell us the accurate predictions
# TP (True Positive) means Actual result = Drug Resistant from model = Drug Resistant from data as well [22]
# TN (True Negative) means Actual result = Not Drug Resistant from model = Not Drug Resistant from data as well [40]
# FP (False Positive) means Actual result = Drug Resistant from model WHEN Not Drug Resistant from data [15]
# FN (False Negative) means Actual result = Not Drug Resistant from model WHEN Drug Resistant from data [8]
```

```python
#Visualizing confusion matrix in a heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True)
plt.xlabel("prediction label")
plt.ylabel("actual label")
plt.show()
```

```python
#Measuring Accuracy
#Accuracy = Correct predictions divided by total observations = TP+TN / TP+TN +FP + FN
#Precision = Ration of correctly predicted +ve observations to the total positive observations = TP / TP + FP (it therefore tells how many patients are actually Drug Resistant)
#Recall = Ratio of correctly predicted +ve observations to ALL observations in the class = TP / TP + FN (how many did the model label as true Drug Resistant)
#F1 score = Weighted average of precision and recall 
```

```python
print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))
print("Precision: ", metrics.precision_score(y_test, y_pred))
print("Recall: ", metrics.recall_score(y_test, y_pred))
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
kfold = KFold(n_splits = 3, random_state = 7)
result = cross_val_score(logr, x_train, y_train, cv= kfold, scoring = "accuracy")
print(result.mean())
print(result)
```

HYPER PARAMETERS TUNING (to optimize / improve results) using Grid Search and Random Search methods.
This shows the best model score and time it takes to make calculations

```python
from sklearn.model_selection import GridSearchCV
dual = [True,False]
max_iter = [100, 110, 120, 130, 140]
param_grid = dict(dual= dual, max_iter = max_iter)
import time
lr = LogisticRegression(penalty = "l2")
grid = GridSearchCV(estimator = lr, param_grid = param_grid, cv =6, n_jobs = 1)
start_time = time.time()
grid_result = grid.fit(x_train, y_train)
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
print("Execution Time " + str((time.time() - start_time))+"ms")
```

```python
from sklearn.model_selection import RandomizedSearchCV
random = RandomizedSearchCV(estimator = lr, param_distributions=param_grid, cv =6, n_jobs = 1)
start_time = time.time()
random_result = random.fit(x_train, y_train)
#summarize the result

print("Best Score: %f using %s" % (random_result.best_score_, random_result.best_params_))
print("Execution Time " + str((time.time() - start_time))+"ms")
```
